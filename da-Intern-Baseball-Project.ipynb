```python
from scipy.stats import zscore
```

```python
# To remove outliers we using zscore method here
z=np.abs(zscore(df))
```

```python
threshold = 3
print(np.where(z<3))
print(df.shape)
```

```python
data = df[(z<3).all(axis=1)]
```

```python
print(df.shape)
print(data.shape)
df=data
print(df.shape)
```

```python
plt.figure(figsize=(20,15))
p=1
for i in df:
    if p<=17:
        plt.subplot(5,4,p)
        sns.regplot(x='W',y=i,data=df,color='r')
        plt.xlabel('W')
        plt.ylabel(i)
    p+=1
plt.show()        
```

1.The number of runs,home runs,doubles,saves,shutouts, and walk are highly positively linearly correlated.
2.stolen Bases,run allowed,earned runs are highly negatively linearly correlated.
3.the remaining features have less to no linear correlation with number of wins.

# Seperating Features and Label

```python
x = df.drop('W',axis=1)
y = df['W']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from sklearn.model_selection import GridSearchCV
```

```python
models = [LinearRegression(),KNeighborsRegressor(),GradientBoostingRegressor()]
max_r2 = 0
for st in range(40,70):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = st)
    for i in models:
        i.fit(x_train,y_train)
        pred = i.predict(x_test)
        r2 = r2_score(y_test,pred)
        if r2>max_r2:
            max_r2 = r2
            final_state = st
            final_model = i
    
    
print('random state:',final_state,'and score:',max_r2,'and model:',final_model)    
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state = 67)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
l_pred = lr.predict(x_test)
print(lr.score(x_test,y_test))
```

```python
print('mean_absolute_error:',mean_absolute_error(y_test,l_pred))
print('mean_squared_error:',mean_squared_error(y_test,l_pred))
print('np.sqrt(mean_absolute_error:',np.sqrt(mean_absolute_error(y_test,l_pred)))
print('r2_score:',r2_score(y_test,l_pred))
```
