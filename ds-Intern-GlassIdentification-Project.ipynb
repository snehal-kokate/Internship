# Seperating Features and Label

```python
x = df.drop('Type',axis = 1)
y = df['Type']
```

# Feture Scaling using Standard Scalarization

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
x = pd.DataFrame(scale.fit_transform(x),columns = x.columns)
x
```

# Checking Variance Inflation Factor(VIF)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif['vif values'] = [variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif['features'] = x.columns
vif
```

drop column which has higher vif value. which is 'RI' column.

```python
x.drop('RI',axis = 1,inplace =True)
```

```python
# again checking VIF to confirm multicoliniarity exist or not
vif = pd.DataFrame()
vif['vif values'] = [variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif['features'] = x.columns
vif
```

# Oversampling

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE()
x1,y1 = sm.fit_resample(x,y)
```

```python
y.value_counts()
```

```python
y1.value_counts() #Balanced Values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier,BaggingClassifier

from sklearn.metrics import classification_report,accuracy_score,confusion_matrix,roc_curve

from sklearn.model_selection import cross_val_score
```

# Split data

```python
x_train,x_test,y_train,y_test = train_test_split(x1,y1,test_size =0.20,random_state = 42)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

# Modeling

```python
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
predrf = rf.predict(x_test)
print(accuracy_score(y_test,predrf))
print(confusion_matrix(y_test,predrf))
print(classification_report(y_test,predrf))
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
predlr = lr.predict(x_test)
print(accuracy_score(y_test,predlr))
print(confusion_matrix(y_test,predlr))
print(classification_report(y_test,predlr))
```

```python
sv = SVC()
sv.fit(x_train,y_train)
predsv = sv.predict(x_test)
print(accuracy_score(y_test,predsv))
print(confusion_matrix(y_test,predsv))
print(classification_report(y_test,predsv))
```

```python
gr = GradientBoostingClassifier()
gr.fit(x_train,y_train)
predgr = gr.predict(x_test)
print(accuracy_score(y_test,predgr))
print(confusion_matrix(y_test,predgr))
print(classification_report(y_test,predgr))
```

```python
bg = BaggingClassifier()
bg.fit(x_train,y_train)
predbg = bg.predict(x_test)
print(accuracy_score(y_test,predbg))
print(confusion_matrix(y_test,predbg))
print(classification_report(y_test,predbg))
```

```python
et = ExtraTreesClassifier()
et.fit(x_train,y_train)
predet = et.predict(x_test)
print(accuracy_score(y_test,predet))
print(confusion_matrix(y_test,predet))
print(classification_report(y_test,predet))
```

# Cross Validation Score

1.cross validation allow us to choose the best performing model.
2.calculating difference between accuracy score and cross validation score
3.Checking cross validation score for each model.
4.performing cross validation on the same oversampled dataset which is used for training and testing models.

```python
score = cross_val_score(rf,x1,y1)
print(score)
print(score.mean())
print('Difference between accuracy score and cv score is:',accuracy_score(y_test,predrf)-score.mean())
```

```python
score = cross_val_score(lr,x1,y1)
print(score)
print(score.mean())
print('Difference between accuracy score and cv score is:',accuracy_score(y_test,predlr)-score.mean())
```

```python
score = cross_val_score(sv,x1,y1)
print(score)
print(score.mean())
print('Difference between accuracy score and cv score is:',accuracy_score(y_test,predsv)-score.mean())
```

```python
score = cross_val_score(gr,x1,y1)
print(score)
print(score.mean())
print('Difference between accuracy score and cv score is:',accuracy_score(y_test,predgr)-score.mean())
```

```python
score = cross_val_score(bg,x1,y1)
print(score)
print(score.mean())
print('Difference between accuracy score and cv score is:',accuracy_score(y_test,predbg)-score.mean())
```

```python
score = cross_val_score(et,x1,y1)
print(score)
print(score.mean())
print('Difference between accuracy score and cv score is:',accuracy_score(y_test,predet)-score.mean())
```

Bagging Classifier is our best model as the difference between accuracy score and cross validation score is least.
